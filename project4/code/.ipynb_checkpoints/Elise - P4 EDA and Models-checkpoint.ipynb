```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
merged_train = pd.read_csv('./input/merged_train.csv')
mapdata = np.loadtxt("./input/mapdata_copyright_openstreetmap_contributors.txt")
```

```python
merged_train.head()
```

```python
merged_train.shape
```

```python
merged_train.columns
```

```python
merged_train.columns = [_.lower() for _ in merged_train.columns]
merged_train.T
```

```python
merged_train['wnvpresent'].value_counts()
```

```python
merged_train['latitude'].sort_values(ascending=False)

#Latitute 41.6 to 42.1
#longitude -87.93 to -87.53
```

```python
wvn_absent = merged_train[['latitude','longitude','wnvpresent']][merged_train['wnvpresent'] == 0]
wvn_present = merged_train[['latitude','longitude','wnvpresent']][merged_train['wnvpresent'] == 1]
```

```python
plt.figure(figsize=(25,15))

origin = [41.6, -88.0]              # lat/long of origin (lower left corner)
upperRight = [42.1, -87.5]          # lat/long of upper right corner

colors = ['red','blue']

# generate plot
plt.imshow(mapdata, cmap=plt.get_cmap('gray'), extent=[origin[1], upperRight[1], origin[0], upperRight[0]])
plt.scatter(x = wvn_present['longitude'], y = wvn_present['latitude'], 
            c = 'red' , s=100)
plt.scatter(x = wvn_absent['longitude'], y = wvn_absent['latitude'], 
            c = 'blue' , s=10)

#plt.scatter(x = merged_train['longitude'], y = merged_train['latitude'], 
            #c = merged_train['wnvpresent'].map(lambda s: colors[s]) , s=20)
plt.show()
#plt.savefig('map.png')
```
